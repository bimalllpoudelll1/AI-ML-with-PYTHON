```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris # plant iris dataset
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")
```

```python
iris = load_iris()
X = iris.data
Y = iris.target

```

```python
Y
```

```python
scaler=StandardScaler()
x_scale=scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train=torch.FloatTensor(X_train).to(device)
X_test=torch.FloatTensor(X_test).to(device)
Y_train=torch.LongTensor(Y_train).to(device)
Y_test=torch.LongTensor(Y_test).to(device)
```

```python
model=nn.Sequential(
    nn.Linear(4,10),
    nn.ReLU(),
    nn.Linear(10,3)
).to(device)
```

```python
criterion=nn.CrossEntropyLoss()#mistake caluclator
optimizer=optim.Adam(model.parameters(),lr=0.01) #weight update
```

```python
# epoch
for epoch in range(100):
    # forward propagation
    outputs=model(X_train)
    loss= criterion(outputs,Y_train)

    # backward propagtion
    # if weights wrong at first ,it cleans memory for another weight
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
model.eval()
with torch.no_grad():
    prediction=model(X_test)
    prediction_classes=torch.max(prediction,1)
    accuracy=(prediction_classes==Y_test).sum().item/len(Y_test)
    print(f'final accuracy:{accuracy*100:,.2f}%')
```
